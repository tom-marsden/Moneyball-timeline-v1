```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
```

```python
# Imports founder features dataset. Edit file path as necessary
features = pd.read_csv('data/features.csv')
features.head()
```

```python
# Extracts successful founder data
success = features[features['success'] == 1]
success.head()
```

```python
# Extracts unsuccessful founder data
fail = features[features['success'] == 0]
fail.head()
```

##Age:

```python
# All histogram plots follow similar method

# Creates two plots in same image
fig, axes = plt.subplots(1, 2, figsize=(10, 4), dpi=300, facecolor='w')

# np.arange controls bin size for histogram
axes[0].hist(success['age'], density = True, bins = np.arange(0,90,5))
# Formatting
axes[0].spines.right.set_visible(False)
axes[0].spines.top.set_visible(False)
axes[0].set_xlim(0,80)
axes[0].set_ylim(0,0.07)
axes[0].set_xlabel('Age in Years')
axes[0].set_ylabel('Probability Density')
axes[0].set_title('Age of Founders on Founding Date (Successful)', pad = 10)

axes[1].hist(fail['age'], density = True, bins = np.arange(0,90,5))
axes[1].spines.right.set_visible(False)
axes[1].spines.top.set_visible(False)
axes[1].set_xlim(0,80)
axes[1].set_ylim(0,0.07)
axes[1].set_xlabel('Age in Years')
axes[1].set_ylabel('Probability Density')
axes[1].set_title('Age of Founders on Founding Date (Unsuccessful)', pad = 10)

plt.subplots_adjust(wspace = 0.5)

# high peaks in centre due to missing data being placed at the mean
```

##Degree subject

```python
len_success = len(success)
len_fail = len(fail)
success_degrees = []
fail_degrees = []
index = []

# i goes over indexes of degree subjects in dataframes
for i in range(3,10):
  # Degree subject features one-hot encoded
  # Sums over relevant column for degree subject and converts into % of total
  success_degrees.append(success.iloc[:,i].sum()/len_success*100)
  fail_degrees.append(fail.iloc[:,i].sum()/len_fail*100)
  # Appends column title (name of degree subject)
  index.append(success.columns[i])

# Plots bar chart
fig, ax = plt.subplots(1, 1, figsize = (8,4), dpi = 300, facecolor='w')
# +- 0.2 and width values set for bars to be correctly spaced
ax.bar(np.arange(len(index)) - 0.2, success_degrees, width = 0.4, align='center', label = 'Successful')
ax.bar(np.arange(len(index)) + 0.2, fail_degrees, width = 0.4, color='r', align='center', label = 'Unsuccessful')

ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)
ax.set_ylabel('Percent of Founders')
ax.set_title('Percent of Founders Studying Each Degree', pad = 10)
ax.set_xticks(np.arange(len(index)), index)
ax.tick_params(labelsize = 8)
ax.yaxis.set_major_formatter(PercentFormatter(decimals = 0))
ax.legend(loc='upper right', ncols=2)

plt.show
```

##Degree Type

```python
# As above but with degree type (PhD, Master's)

len_success = len(success)
len_fail = len(fail)
success_degrees = []
index = []
fail_degrees = []

for i in range(1,3):
  success_degrees.append(success.iloc[:,i].sum()/len_success*100)
  fail_degrees.append(fail.iloc[:,i].sum()/len_fail*100)
  index.append(success.columns[i])

fig, ax = plt.subplots(1, 1, figsize = (8,4), dpi = 300, facecolor='w')
ax.bar([0.2, 0.5], success_degrees, width = 0.1, align='center', label = 'Successful')
ax.bar([0.3, 0.6], fail_degrees, width = 0.1, color='r', align='center', label = 'Unsuccessful')

ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)
ax.set_ylabel('Percent of Founders')
ax.set_title('Percent of Founders with Each Type of Degree', pad = 10)
ax.set_xticks([0.25, 0.55], index)
ax.yaxis.set_major_formatter(PercentFormatter(decimals = 0))
ax.legend(loc='upper left', ncols=2)

plt.show
```
